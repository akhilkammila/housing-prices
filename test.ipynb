```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Create sample data
np.random.seed(42)
categories = ['A', 'B', 'C', 'D']
n_samples = 30

data = {
    'Category': np.repeat(categories, n_samples),
    'Value': np.concatenate([
        np.random.normal(10, 2, n_samples),  # Group A
        np.random.normal(12, 1.5, n_samples),  # Group B
        np.random.normal(8, 2, n_samples),  # Group C
        np.random.normal(15, 1, n_samples)   # Group D
    ])
}
df = pd.DataFrame(data)

# Create figure with two subplots
plt.figure(figsize=(12, 5))

# Pointplot
plt.subplot(1, 2, 1)
sns.pointplot(data=df, x='Category', y='Value', color='blue')
plt.title('Pointplot: Shows Mean & Confidence Interval')
plt.ylim(5, 17)

# Stripplot
plt.subplot(1, 2, 2)
sns.stripplot(data=df, x='Category', y='Value', color='blue', alpha=0.5)
plt.title('Stripplot: Shows Individual Data Points')
plt.ylim(5, 17)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Demonstration of stacking behavior
plt.plot(x, y1, 'b-', label='sin(x)')
plt.plot(x, y2, 'r--', label='cos(x)')  # Stacks on same axes
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Clear everything to start fresh
plt.close('all')

# Test 1: What happens with gcf() on empty state
fig = plt.gcf()
print("Default figure size (inches):", fig.get_size_inches())
print("Default figure DPI:", fig.get_dpi())
print("Default figure size (pixels):", 
      fig.get_size_inches() * fig.get_dpi())

# Get default margins and spacing
ax = plt.gca()
print("\nDefault margins:")
print("Left margin:", ax.get_position().x0)
print("Right margin:", 1 - ax.get_position().x1)
print("Bottom margin:", ax.get_position().y0)
print("Top margin:", 1 - ax.get_position().y1)

# Show what the default layout looks like with some content
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x))
ax.set_title("Default Figure and Axes")
plt.show()

# Test 2: Create new figure and check if gca() creates axes
plt.close('all')
fig = plt.figure()
print("\nBefore gca(): Number of axes in figure:", len(fig.axes))
ax = plt.gca()
print("After gca(): Number of axes in figure:", len(fig.axes))

# Test 3: Show default subplot parameters
print("\nDefault subplot parameters:")
print(plt.rcParams['figure.subplot.left'])
print(plt.rcParams['figure.subplot.right'])
print(plt.rcParams['figure.subplot.bottom'])
print(plt.rcParams['figure.subplot.top'])

# Show that repeated calls get the same figure/axes
plt.close('all')
fig1 = plt.gcf()
fig2 = plt.gcf()
print("\nRepeated gcf() calls return same figure:", fig1 == fig2)

ax1 = plt.gca()
ax2 = plt.gca()
print("Repeated gca() calls return same axes:", ax1 == ax2)
```
